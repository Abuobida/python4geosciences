```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)

x = np.linspace(0, np.pi, 300)
n_values = np.arange(1, 6)

# y = np.sin( n[np.newaxis, :] * x[:, np.newaxis] )
# plt.plot(x, y)

linestyles = ['-', '--', ':', '-.', '-']
colors = ['r', 'y', 'orange', 'g', 'b']

lines = []
for n, ls, color in zip(n_values, linestyles, colors):
    label = 'mode={}'.format(n)
    line = ax.plot(x, np.sin(n * x), ls=ls, color=color, label=label)
    lines.append(line)

plt.legend(loc='lower left')

ax.set_xticks([0, np.pi/2, np.pi]);
ax.set_xticklabels(['0', '$\pi$/2', '$\pi$'], fontsize=14);
```
