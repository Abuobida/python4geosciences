```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from scipy.special import j0, j1, jn
```

```python
x = np.linspace(0, 20, 201)
plt.plot(x, j0(x), '-r')
```

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)

x = np.linspace(-20, 20, 201)
y = np.linspace(-20, 20, 201)
x, y = np.meshgrid(x, y)

# y, x = np.mgrid[-20:20:201j, -20:20:101j]

r = np.sqrt(x**2 + y**2)
theta = np.arctan2(y, x)

z = j0(r)
zmask = np.ma.masked_where( (theta < 0.4) & (theta > -0.4), z)

ax.pcolormesh(x, y, zmask)
```
